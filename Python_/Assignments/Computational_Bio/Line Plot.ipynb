####  Q1> Here are two lists(Year and GDP).On a basis of that create a line plot with proper labels and figure.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year=[1950,1960,1970,1980,1990]
GDP=[50.8,65.9,75.8,92.6,125.8]

plt.figure(figsize=(5,4))
plt.plot(Year,GDP)
plt.title("Year Vs GDP")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.grid()
plt.show()
```
