```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the tips dataset

tips=sns.load_dataset('tips')
tips
```

#### Q1> Draw a scatter plot on a basis of above data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot")
plt.show()
```

#### Q2> Draw a boxplot on the basis of above data set.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box Plot')
plt.show()
```

#### Q3> Draw a bar plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot')
plt.show()
```

#### Q4> Draw a violin plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin Plot")
plt.show()
```

#### Q5> Draw a line plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title("Line PLot")
plt.show()
```

#### Q6> Draw a cat plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.catplot(x="total_bill",y="tip",data=tips)
plt.title("Cat Plot")
plt.show()
```

#### Q7> Draw a Swarm Plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

sns.stripplot(x="total_bill", y="tip", data=tips, jitter=True, size=4)

plt.title("Strip Plot with Jitter")
plt.show()
```
